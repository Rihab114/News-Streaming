```python
#define variables
API_KEY = ''
SERVERS = ''
TOPIC = 'kafka-streaming'
```

```python
#define the consumer
consumer = KafkaConsumer(TOPIC, bootstrap_servers = SERVERS) 
```

```python
#Initialize S3 file system 
s3 = S3FileSystem()
```

```python
#define functions to send data to s3
def write_to_s3(lines):
    file_content = '\n'.join(lines)
    with s3.open('', 'a') as f:
        f.write(file_content)
```

```python
lines_buffer = []
for message in consumer:
    line = message.value.decode('utf-8')
    lines_buffer.append(line)
    if len(lines_buffer) == 3:
        write_to_s3(lines_buffer)
        lines_buffer = []
```
